```python
# Write a Python program using Pandas to create a DataFrame from the given data and
# find the total revenue and average price of items sold in each store.

data = {
    'Store': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Item': ['Apple', 'Banana', 'Orange', 'Grape', 'Apple', 'Banana', 'Orange', 'Grape'],
    'Price': [50, 20, 30, 60, 55, 22, 33, 65],
    'Quantity': [10, 12, 15, 16, 20, 25, 30, 35]
}

df = pd.DataFrame(data)
df['Revenue'] = df['Price'] * df['Quantity']
store_revenue = df.groupby('Store')['Revenue'].sum()
total_revenue = store_revenue.sum()
avg_price = df.groupby(['Store','Item'])['Price'].mean().reset_index()
print("Total Revenue per store:")
print(store_revenue)
print(f"Total Revenue:{total_revenue}")
print(f"Average price per store per item:")
print(avg_price)



```

```python
# Write a Python program using Pandas to create a DataFrame from the given data and
# calculate the total amount spent by each customer.

data = {
    'Customer': ['Alice', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Bob'],
    'Item': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Pen', 'Pencil', 'Notebook', 'Eraser'],
    'Price': [10, 5, 50, 20, 10, 5, 50, 20],
    'Quantity': [3, 4, 2, 5, 10, 6, 1, 2]
}
df = pd.DataFrame(data)
df['Amount'] = df['Price'] * df['Quantity']
df['Total_Spent'] = df['Price'] * df['Quantity']
total_amt_spent_per_customer = df.groupby('Customer')['Total_Spent'].sum().reset_index()
#total_amt_spent_per_customer = df.groupby('Customer').agg({'Total_Spent': 'sum'}).reset_index()
print(total_amt_spent_per_customer)
```

```python
# Write a Python program using Pandas to create a DataFrame from the given data and
# find the month with the highest average temperature.



data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Temperature': [20, 22, 25, 27, 30, 32, 35, 34, 30, 28, 24, 21]
}
df = pd.DataFrame(data)
highest_temperature = df.loc[df['Temperature'].idxmax()]['Month']

print(highest_temperature)
```

```python
# Write a Python program using Pandas to create a DataFrame from the given data and
# find the total amount spent per item, as well as the total amount spent by all customers.

data = {
    'Customer': ['Alice', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Bob'],
    'Item': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Pen', 'Pencil', 'Notebook', 'Eraser'],
    'Price': [10, 5, 50, 20, 10, 5, 50, 20],
    'Quantity': [3, 4, 2, 5, 10, 6, 1, 2]
}
df = pd.DataFrame(data)

df['Total_Amount_Spent'] = df['Price'] * df['Quantity']
print(df)
total_amt_spent_per_item = df.groupby('Item')[['Total_Amount_Spent']].sum()
total_amt_spent= total_amt_spent_per_item['Total_Amount_Spent'].sum()
print(total_amt_spent_per_item)
print(f"Total Amount Spent:{total_amt_spent}")

```

```python
# Write a Python program using Pandas to create a DataFrame from the given data and
# calculate the monthly average temperature difference between two cities.

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'City_A_Temp': [20, 22, 25, 27, 30, 32, 35, 34, 30, 28, 24, 21],
    'City_B_Temp': [10, 12, 15, 17, 20, 22, 25, 24, 20, 18, 14, 11]
}

df = pd.DataFrame(data)
print(df)
avg_diff = (abs(df['City_A_Temp'] - df['City_B_Temp'])).mean()
print(avg_diff)
```
